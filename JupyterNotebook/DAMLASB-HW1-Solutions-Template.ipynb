```python
def character_count(sampletext):
    sampletext = sampletext.lower()
    sampletext = sampletext.replace(" ", "")
    dictnry = {}
    for i in sampletext:
        if i not in dictnry:
            dictnry.setdefault(i,1)
        else:
            dictnry[i] = dictnry[i] + 1
    dictnry = sorted(dictnry.items())
    for i in dictnry:
        print(str(i[0])+' '+ str(i[1]))

character_count('ThiS is String with Upper and lower case Letters')
```
